```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Sales Records.csv')

df['Order Date'] = pd.to_datetime(df['Order Date'])

df['Profit Margin'] = df['Total Profit'] / df['Total Revenue']

df.sort_values(by='Order Date', inplace=True)

top_items_revenue = df.groupby('Item Type')['Total Revenue'].sum().head(10)
plt.figure(figsize=(10, 6))
top_items_revenue.plot(kind='bar', color='skyblue')
plt.title('Top Selling Items by Total Revenue')
plt.xlabel('Item Type')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45, ha='right')
plt.show()

sns.set(style='whitegrid', palette='pastel')

plt.figure(figsize=(10, 6))
top_items_units_sold = df.groupby('Item Type')['Units Sold'].sum().head(10)
plt.bar(top_items_units_sold.index, top_items_units_sold, color='Green')
plt.title('Top Selling Items by Total Units Sold')
plt.xlabel('Item Type')
plt.ylabel('Units Sold')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Sales Records.csv')

df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')

df = df.dropna(subset=['Order Date'])


df_aggregated = df.groupby('Order Date')['Total Profit'].mean().reset_index()

df_aggregated.sort_values(by='Order Date', inplace=True)

plt.figure(figsize=(12, 6))
plt.plot(df_aggregated['Order Date'], df_aggregated['Total Profit'], linestyle='-', color='green', label='Mean Profit Margin')
plt.title('Trend of Mean Profit Over Time')
plt.xlabel('Order Date')
plt.ylabel('Profit')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.preprocessing import LabelEncoder
import pandas as pd

data = pd.read_csv('Sales Records.csv')

target_columns = ['Units Sold', 'Unit Price','Unit Cost', 'Total Revenue', 'Total Cost', 'Total Profit']  # Replace with your actual target column names

for target in target_columns:
    X = data[['Region', 'Country', 'Item Type', 'Sales Channel', 'Order Priority',
              'Order Date', 'Order ID', 'Ship Date', 'Units Sold', 'Unit Price',
              'Unit Cost', 'Total Revenue', 'Total Cost', 'Total Profit']]
    
    y = data[target]

    label_encoder = LabelEncoder()
    
    for column in X.select_dtypes(include=['object']).columns:
        X[column] = label_encoder.fit_transform(X[column])

    x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error  
from sklearn.metrics import r2_score

linreg = LinearRegression()
linreg.fit(x_train, y_train)

    # Make predictions
y_pred = linreg.predict(x_test)

    # Evaluate the model using R-squared
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test,y_pred)

print(f"R-squared for {target}: {r2}")
print(f"Mean Squared Error: {mse}") 
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split

np.random.seed(42)
x_train, x_test, y_train, y_test = np.random.rand(50, 5), np.random.rand(10, 5), np.random.rand(50), np.random.rand(10)


models = [
    ("Linear Regression", LinearRegression()),
    ("Random Forest Regression", RandomForestClassifier()),
    ("KNN Regression", KNeighborsRegressor()),
    ("Decision Tree Regression", DecisionTreeRegressor()),
    ("Gradient Boosting Regression", GradientBoostingRegressor())
]

# Loop through models
for model_name, model in models:
    # Train the model
    model.fit(x_train, y_train)
    
    # Make predictions
    y_pred = model.predict(x_test)
    
    # Display regression metrics
    print(f"\nMetrics for {model_name}:")
    print("Mean Absolute Error:", mean_absolute_error(y_test, y_pred))
    print("Mean Squared Error:", mean_squared_error(y_test, y_pred))
    print("R-squared:", r2_score(y_test, y_pred))

    
```

```python
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming x_train, y_train, x_test, y_test are already defined

# Initialize regression models
models = [
    ('Random Forest Regressor', RandomForestRegressor()),
    ('Decision Tree Regressor', DecisionTreeRegressor()),
    ('Linear Regression', LinearRegression()),
    ('KNN Regressor', KNeighborsRegressor()),
    ('Gradient Boosting Regressor', GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42))
]

mse_scores = []
r2_scores = []

# Train and evaluate each model
for model_name, model in models:
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    
    mse_scores.append(mse)
    r2_scores.append(r2)
    
    print(f"{model_name} - Mean Squared Error: {mse}, R-squared: {r2}")

# Plotting Mean Squared Error
plt.figure(figsize=(10, 6))
sns.barplot(x=[model[0] for model in models], y=mse_scores, palette='magma').set_title('Model Comparison (Mean Squared Error)')
plt.xticks(rotation=45)
plt.ylabel('Mean Squared Error')
plt.show()

# Plotting R-squared;l-;,m        
plt.figure(figsize=(10, 6))
sns.barplot(x=[model[0] for model in models], y=r2_scores, palette='magma').set_title('Model Comparison (R-squared)')
plt.xticks(rotation=45)
plt.ylabel('R-squared')
plt.show()
```
